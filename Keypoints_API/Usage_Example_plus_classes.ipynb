# API Usage

```python
%matplotlib inline
import skimage.io as io
import skimage.transform as transform
import matplotlib.pyplot as plt

example = io.imread('examples/face.jpg')
plt.imshow(example)
plt.axis('off')
plt.show()
```

```python
from get_keypoints import *

keypoints = get_raw_keypoints(example)
keypoints.shape
```

```python
result = transform_img_with_keypoints(example)

result = transform.resize(result, example.shape, anti_aliasing=True)
```

```python
dpi = 80
height, width, nbands = example.shape
figsize = width / float(dpi), height / float(dpi)

fig = plt.figure(figsize=figsize)
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('off')
ax.imshow(result, interpolation='nearest')

fig.savefig('examples/keypoints.png', dpi=dpi, transparent=True)
plt.show()
```
