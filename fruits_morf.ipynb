```python
import numpy as np
import skimage
from skimage import io
from skimage import transform
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
banana = io.imread('photos/banana1.png')
plt.imshow(banana)

plt.show()
```

```python
src = np.array([
    (10, 100),
    (450, 350),
    (800, 200)
])

dst = np.array([
    (100, 100),
    (450, 400),
    (800, 250)
])

plt.imshow(banana)
plt.scatter(*list(zip(*src)), label='src')
plt.scatter(*list(zip(*dst)), label='dst')
plt.legend()
plt.show()
```

```python
from skimage.transform import warp

t = transform.estimate_transform('affine', src, dst)
warp_banana = warp(banana, inverse_map=t.inverse)

plt.imshow(warp_banana)
plt.scatter(*list(zip(*src)), label='src')
plt.scatter(*list(zip(*dst)), label='dst')
plt.legend()
plt.show()
```

```python
warp_banana = warp(banana, inverse_map=t.inverse, output_shape=(500, 1000))
plt.imshow(warp_banana)
plt.scatter(*list(zip(*src)), label='src')
plt.scatter(*list(zip(*dst)), label='dst')
plt.legend()
plt.show()
```
